```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

dfConst = pd.read_csv('C:/Users/amand/Downloads/CTBA/CTBA/M3/ConstructionTimeSeriesDatav2.csv')

x  = dfConst['Month-Year']
y = dfConst['Private Construction']

fig,ax = plt.subplots()
#Private Construction Graph
ax.plot(x,y,label='Private Contruction Spending',color='k')
ax.yaxis.set_label_text('Private Contruction Spending',fontsize=14) #y-axis label
ax.xaxis.set_label_text('Month-Year',fontsize=14) #x-axis label
ax.xaxis.set_tick_params(labelsize=14)
ax.yaxis.set_tick_params(labelsize=14)
#Public Construction Graph
ax2=ax.twinx()
ax2.plot(x,dfConst['Public Construction'],label='Public Contruction Spending',color='r')
ax2.yaxis.set_label_text('Public Contruction Spending',fontsize=14) #y-axis label
ax2.yaxis.set_tick_params(labelsize=14)
#Removing the lines at the top and right
ax.spines['top'].set_visible(False)
ax2.spines['top'].set_visible(False)
#Limiting the number of ticks and labels
ax.set_xticks(x[::12])
ax.set_xticklabels(x[::12])
ax.set_ylim(min(y)-10000,max(y)+30000)
#create a colored background
ax.axvspan(63,145,ymin=0,ymax=120000,facecolor='k',alpha=0.1) #alpha determines transparency
ax.text(110,27000,'Recession and Recovery',fontsize=12,verticalalignment='center',horizontalalignment='center')
ax.legend(loc=2)
ax2.legend(loc=1)
fig.set_size_inches(14,6)
fig.suptitle('Total Construction per Year',fontsize='25')
plt.show()
fig.savefig('lineplot.jpg',dpi=600)
```

The graph is highlighting how construction spending changed around the time of the recession. Prior to that time, the rates of change in spending between the public and private sectors remained relatively similar but during the recession, public spending stayed steady while the rate of private spending dropped dramatically.
